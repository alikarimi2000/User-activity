```python
import sqlite3

conect= sqlite3.connect("dkcup_test.db")

cursor = conect.cursor()

query = """
WITH ranked_logs AS (
    SELECT 
        user_id, 
        date, 
        log,
        ROW_NUMBER() OVER (PARTITION BY user_id ORDER BY date) AS rn
    FROM 
        users_log
),
grouped_logs AS (
    SELECT 
        user_id, 
        date, 
        log, 
        date(rn - ROW_NUMBER() OVER (PARTITION BY user_id, log ORDER BY date)) AS grp
    FROM 
        ranked_logs
),
aggregated_logs AS (
    SELECT 
        user_id,
        MIN(date) AS start_date,
        MAX(date) AS end_date,
        log,
        COUNT(*) AS length
    FROM 
        grouped_logs
    GROUP BY 
        user_id, log, grp
)
SELECT 
    user_id,
    start_date,
    end_date,
    log,
    length
FROM 
    aggregated_logs
ORDER BY 
    user_id, start_date;


"""

cursor.execute(query)
rows = cursor.fetchall()

```
